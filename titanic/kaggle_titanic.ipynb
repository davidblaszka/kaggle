```python
def plot_distribution( df , var , target , **kwargs ):
    row = kwargs.get( 'row' , None )
    col = kwargs.get( 'col' , None )
    facet = sns.FacetGrid( df , hue=target , aspect=4 , row = row , col = col )
    facet.map( sns.kdeplot , var , shade= True )
    facet.set( xlim=( 0 , df[ var ].max() ) )
    facet.add_legend()
```

```python
# Plot distributions of Age of passangers who survived or did not survive
plot_distribution( titanic_df , var = 'Age' , target = 'Survived' , row = 'Sex' )
```

```python
title_dict = {'Mr': 'mr',
                'Mrs': 'mrs',
                'Miss': 'miss',
                'Don': 'royalty',
                'Master': 'royalty',
                'Dr': 'honorary',
                'Mme': 'mrs',
                'Major': 'officer',
                'Lady': 'royalty',
                'Sir': 'royalty',
                'Rev': 'honorary',
                'Mlle': 'miss',
                'Col': 'officer',
                'Capt': 'officer',
                'the Countess': 'royalty',
                'Jonkheer': 'royalty',
                'Dona': 'royality',
                'Ms': 'mrs',
               }
```

```python
title = pd.DataFrame()
# we extract the title from each name
title[ 'Title' ] = full_df[ 'Name' ].map( lambda name: name.split( ',' )[1].split( '.' )[0].strip() )
```

```python
# we map each title
title['Title'] = title.Title.map(title_dict)
df_dum_title = pd.get_dummies(title.Title).drop('honorary', axis=1)
df_dum_title.head()
```

Make dummies of sex

```python
df_dum_sex = pd.get_dummies(full_df.Sex).drop('male', axis=1)
```

Make dummies of embarked

```python
df_dum_embarked = pd.get_dummies(full_df.Embarked).drop('C',axis=1)
df_dum_embarked.columns = map(lambda x: 'embarked_' + str(x), df_dum_embarked.columns)
```

cabin dummies

```python
df_dum_cabin = pd.DataFrame()

# replacing missing cabins with U (for Uknown)
df_dum_cabin['Cabin'] = full_df.Cabin.fillna('U')

# mapping each Cabin value with the cabin letter
df_dum_cabin['Cabin'] = df_dum_cabin['Cabin'].map(lambda c : c[0])

# dummy encoding ...
df_dum_cabin = pd.get_dummies(df_dum_cabin['Cabin'] , prefix = 'Cabin').drop('Cabin_T', axis=1)

df_dum_cabin.head()
```

```python
df_dum_pclass = pd.get_dummies(full_df.Pclass)
df_dum_pclass.columns = map(lambda x: str(x) + '_class', df_dum_pclass.columns)
```

```python
# a function that extracts each prefix of the ticket, returns 'XXX' if no prefix (i.e the ticket is a digit)
def cleanTicket( ticket ):
    ticket = ticket.replace( '.' , '' )
    ticket = ticket.replace( '/' , '' )
    ticket = ticket.split()
    ticket = map( lambda t : t.strip() , ticket )
    ticket = list(filter( lambda t : not t.isdigit() , ticket ))
    if len( ticket ) > 0:
        return ticket[0]
    else: 
        return 'XXX'

ticket = pd.DataFrame()

# Extracting dummy variables from tickets:
ticket[ 'ticket' ] = full_df[ 'Ticket' ].map( cleanTicket )
df_dum_ticket = pd.get_dummies( ticket[ 'ticket' ] , prefix = 'ticket' ).drop('ticket_XXX', axis=1)

df_dum_ticket.head()
```

```python
df_dum_sibSp = pd.get_dummies(full_df.SibSp)
df_dum_sibSp.columns = map(lambda x: 'sibSp_' + str(x), df_dum_sibSp.columns)
```

```python
df_dum_parch = pd.get_dummies(full_df.Parch)
df_dum_parch.columns = map(lambda x: 'parch_' + str(x), df_dum_parch.columns)
```

Let's collect all these dummies together

```python
engineered_df = full_df
engineered_df = pd.concat([full_df, df_dum_title, df_dum_sex, 
                           df_dum_embarked, df_dum_cabin, df_dum_pclass,
                           df_dum_ticket, df_dum_sibSp, df_dum_parch], axis=1)
engineered_df.drop(['Sex', 'Name', 'Embarked', 'Cabin', 'Pclass', 'Ticket', 'SibSp', 'Parch'], axis=1, inplace=True)
engineered_df.head().T
```

```python
engineered_df.info()
```

Since Age is still missing an 1/8 of the data, let's try and predict the missing values using regression.


```python
engineered_df['age_missing'] = np.where(engineered_df['Age'].isnull(), 1,0)
```

```python
# Fill missing values of Age with the average of Age (mean)
engineered_df['Age'] = engineered_df.Age.fillna(engineered_df.Age.mean())
engineered_df['Fare'] = engineered_df.Fare.fillna(engineered_df.Fare.mean())
```

### Modeling

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
X = engineered_df.drop(['Survived', 'PassengerId'], axis=1)[:891]
y = engineered_df['Survived'][:891]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
# Score the model
print (rfc.score(X_train, y_train) , rfc.score( X_test , y_test))
```

```python
def plot_feature_importance(model, X, num_features=10):
    '''Plot feature importance. Taken from Galvanize lecture code.'''
    feature_importance = model.feature_importances_
    # normalize importances by max importance
    feature_importance = 100.0 * (feature_importance / feature_importance.max())
    sorted_idx = np.argsort(feature_importance)
    ind = np.arange(sorted_idx.shape[0])
    
    # Display num_features
    ind = ind[-num_features:]
    feature_importance = (feature_importance[sorted_idx])[-num_features:]
    feature_names = (X.columns[sorted_idx])[-num_features:]
    
    plt.barh(ind, feature_importance, align='center')
    plt.yticks(ind, feature_names)
    plt.xlabel('Relative Importance')
    plt.title('Variable Importance')
```

```python
plot_feature_importance(rfc, X_train, num_features=15)
```

### Fine tune the models

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_grid = {
    'max_depth': [4, 8, None],
    'max_features': ['sqrt', 'log2', None],
    'min_samples_split': [2, 4],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True], # Mandatory with oob_score=True
    'n_estimators': [50, 100, 200, 400],
    'random_state': [42],
    'oob_score': [True],
    'n_jobs': [-1]
    }
```

```python
rf_grid_cv = GridSearchCV(RandomForestClassifier(),
                             rf_grid,
                             n_jobs=-1,
                             verbose=True,
                             scoring='roc_auc')
```

```python
rf_grid_cv.fit(X_train, y_train)
```
